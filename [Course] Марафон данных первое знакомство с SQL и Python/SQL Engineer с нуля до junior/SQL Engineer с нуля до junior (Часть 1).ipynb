### Неструктурированные данные не имеют заранее определенной модели данных и хранятся в своих собственных форматах (так называемых «исходных» форматах). Это могут быть аудио (WAV, MP3, OGG и т. Д.) Или видеофайлы (MP4, WMV и т. Д.), PDF-документы, изображения (JPEG, PNG и т. Д.), Электронные письма, сообщения в социальных сетях, данные датчиков и т. д.

### **Транзакция** (в базах данных) — это логическая единица работы, которая включает в себя одну или несколько операций, выполняемых над данными в базе. Транзакция может включать такие операции, как вставка, обновление или удаление данных.

### **Транзакция должна быть выполнена полностью или не выполнена вообще** - это необходимо для обеспечения целостности и согласованности данных.

### В рамках одной транзакции система должна выполнить ряд проверок и действий для исключения необоснованного списания денежных средств с карты или изменения остатка товара.

### Аналитические БД, чаще всего, на вход получают данные из транзакционных баз и обрабатывают их порциями (за час, день). После получения новых данных происходит перерасчет агрегированных показателей.
